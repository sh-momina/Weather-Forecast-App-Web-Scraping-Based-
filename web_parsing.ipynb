```python
from bs4 import BeautifulSoup
import requests
```

```python
responce = requests.get("https://wise.com/gb/currency-converter/usd-to-pkr-rate")
```

```python
content = responce.text
```

```python
content
```

```python
soup = BeautifulSoup(content, "html.parser")
```

```python
data = soup.find_all("p")
for i in data:
    print(i.get_text())
```

```python
dollar_price = soup.find(id="target-input")
```

```python
dollar_price
```

```python
dollar_price = dollar_price.get("value")
```
