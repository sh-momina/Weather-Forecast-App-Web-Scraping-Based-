```python
name = {}
for i in a:
    b = i.split()
    name[i] = ""
    for j in b:
        name[i] += j + "_"
name

countries_link = []
for key in name:
    if name[key].endswith("_"):
        name[key] = name[key][:-1]
        countries_link.append("https://weather.metoffice.gov.uk/world/"+name[key].lower())
countries_link
```

```python
cities = []
cities_weather = []
cities_link = []

for link in countries_link:
    cities_names_list = []
    cities_weather_list = []
    cities_link_list = []
    
    url = link
    # print(link)
    r = requests.get(url)
    soup = BeautifulSoup(r.text, "html.parser")
    data = soup.find_all("ul", class_ = "link-group-list")
    # print(data)
    for i in data:
        data2 = i.find_all("span", class_ = "link-text")
        weather_data = i.find_all("span", class_= "temp")
        cities_link_data = i.find_all("href")
        
        for i in data2:
            # print(i.text)
            cities_names_list.append(i.text)
        cities.append(cities_names_list)
        # print("_"*100)
        
        for i in weather_data:
            # print(i.text)
            cities_weather_list.append(i.text)
        cities_weather.append(cities_weather_list)
        # print("_"*100)
        
        for i in data:
            tags = i.find_all("a", href=True)
            for j in tags:
                cities_link_list.append(j["href"])
        cities_link.append(cities_link_list)
        # print("_"*100)
        
# cities_weather
# cities
# cities_link
```
